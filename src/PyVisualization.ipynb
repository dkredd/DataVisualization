```python
%matplotlib inline
df.plot(x="Rank",y=["P75th","Median","P25th"])
```

```python
df.Median.plot(kind='hist')
```

```python
top_5 = df.sort_values(by='Median',ascending=False).head()
top_5
```

```python
top_5.plot(x="Major",y="Median",kind="bar",rot=20,fontsize=6)
```

```python
df[df['Median'] > 60000].sort_values('Median').plot(x="Major",y=["P25th","Median","P75th"],kind="bar")
```

```python
df.plot(x="Median",y="Unemployment_rate",kind="scatter")
```

```python
cat_totals = df.groupby("Major_category")["Total"].sum().sort_values()
cat_totals
```

```python
cat_totals.plot(kind="barh")
```

```python
small_cat_totals = cat_totals[cat_totals < 100000]
large_cat_totals = cat_totals[cat_totals > 100000]
```

```python
small_cat_series = pd.Series([small_cat_totals.sum()],index=["other"])
large_cat_totals.append(small_cat_series).plot(kind="pie",label="")
```
